```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# Load the FLAN-T5 model
model_name = "google/flan-t5-base"
original_model = AutoModelForSeq2SeqLM.from_pretrained(model_name, torch_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Example: Generating a response
input_text = "Summarize: The Eiffel Tower is a famous landmark in Paris, built in 1889."
tokens = tokenizer(input_text, return_tensors="pt").to("cuda")

# Generate output
output_tokens = original_model.generate(**tokens)
output_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)

print("Generated Summary:", output_text)
```
